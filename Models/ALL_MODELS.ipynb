```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error,r2_score,f1_score
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
d={}
for i in range(1,76):
  data=pd.read_excel(str(i)+'.xlsx',index_col='Year')
  data1=data.drop(['District','District ID','Latitude','Longitude'],axis=1)
  data1=pd.get_dummies(data1)
  data1.index = pd.to_datetime(data1.index, format='%Y')
  train_data = data1.loc[:'2011']
  test_data = data1.loc['2012':]
  model = sm.tsa.arima.ARIMA(train_data['Groundwater Recharge'], exog=train_data[['Slope', 'Temperature','Rainfall']], order=(1, 1, 1))
  result = model.fit()
  y_pred = result.forecast(len(test_data), exog=test_data[['Slope', 'Temperature','Rainfall']])
  y_pred_train = result.forecast(len(train_data), exog=train_data[['Slope', 'Temperature','Rainfall']])

  mse_test = mean_squared_error(test_data['Groundwater Recharge'], y_pred)
  mse_train = mean_squared_error(train_data['Groundwater Recharge'], y_pred_train)

  r2_s=r2_score(test_data['Groundwater Recharge'],y_pred)
  # f1_s=f1_score(test_data['Groundwater Recharge'],y_pred)
  d[i]=(mse_train,mse_test,r2_s)
```

```python
df=pd.DataFrame.from_dict(d,orient='index')
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
df = pd.read_excel('1.xlsx')
le=LabelEncoder()
data1=df.drop(['District','District ID','Latitude','Longitude','Soil Type'],axis=1)
```

```python
data1['LULC']=le.fit_transform(data1['LULC'])
# data1['Soil Type']=le.fit_transform(data1['Soil Type'])
```

```python
X = data1[['Rainfall', 'Temperature', 'LULC']].values
y = data1[['Groundwater Recharge']].values

# Scaling the input data
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)

# Splitting the data into training and testing sets
train_size = int(len(X) * 0.67)
test_size = len(X) - train_size
X_train, X_test = X[0:train_size,:], X[train_size:len(X),:]
y_train, y_test = y[0:train_size,:], y[train_size:len(y),:]

# Reshaping the data to fit the LSTM model
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Defining the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(1, 3)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Training the model
model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=2)

# Evaluating the model on the test data
train_score = model.evaluate(X_train, y_train, verbose=0)
test_score = model.evaluate(X_test, y_test, verbose=0)
print('Train Score: %.2f MSE (%.2f RMSE)' % (train_score, np.sqrt(train_score)))
print('Test Score: %.2f MSE (%.2f RMSE)' % (test_score, np.sqrt(test_score)))
```
