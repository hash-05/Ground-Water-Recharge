```python
# Define the model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(33, input_shape=(4,), activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(512, activation='leaky_relu'),
    tf.keras.layers.Dense(1024, activation='leaky_relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
# Train the model
model.fit(X_train, y_train, epochs=200, batch_size=4)
```

```python
# Evaluate the model on the test set
mse_test = model.evaluate(X_test, y_test)
mse_train=model.evaluate(X_train,y_train)
y_pred=model.predict(X_test)

print(f'Mean squared error on test set:{mse_test} train set {mse_train}')
print("R2 square",r2_score(y_test,y_pred))
```
